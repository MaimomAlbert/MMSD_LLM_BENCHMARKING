```python
import pandas as pd

df1 = pd.read_json("gpt_predictions/llama32_11B_osint_batch_3.json")
df2 = pd.read_json("gpt_predictions/internvl25_osint_batch_3.json")
df3 = pd.read_json("gpt_predictions/llava16_mistral_osint_batch_3.json")
df4 = pd.read_json("gpt_predictions/llava16_vicuna_osint_batch_3.json")
df5 = pd.read_json("gpt_predictions/idefics2_9B_osint_batch_3.json")
```

```python
a = pd.merge(df1, df2, on = "image_id", how="inner")
a = a.drop(columns=["text_y", "subreddit_y"])
a.columns = ["image_id", "text", "subreddit", "llama32_11B", "intervl25_8B"]
a.shape
```

```python
b = pd.merge(a, df3, on="image_id")
b = b.drop(columns=["text_y", "subreddit_y"])
b.columns = ["image_id", "text", "subreddit", "llama32_11B", "intervl25_8B", "Llava16_mistral_7B"]
b
```

```python
c = pd.merge(b, df4, on="image_id")
c = c.drop(columns=["text_y", "subreddit_y"])
c.columns = ["image_id", "text", "subreddit", "llama32_11B", "intervl25_8B", "llava16_mistral_7B", "llava16_vicuna_7B"]
c
```

```python
d = pd.merge(c, df5, on="image_id")
d = d.drop(columns=["text_y", "subreddit_y"])
d.columns = ["image_id", "text", "subreddit", "llama32_11B", "intervl25_8B", "llava16_mistral_7B", "llava16_vicuna_7B", "idefics2_9B"]
d
```
