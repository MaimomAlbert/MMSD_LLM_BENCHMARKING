```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/sinngamkhaidem/Developer/mllm-based-mmsd-osint/mllm_finetune/gpt_predictions/batch_1_2_3.csv")
```

```python
df.shape
```

```python
df_pos = df[df["major_label"] == 1]
df_neg = df[df["major_label"] == 0]
```

```python
df_new =  pd.concat([df_pos, df_neg])
```

```python
df_new.shape
```

```python
value_counts = df_new['num_models'].value_counts().sort_index()

# Plot bar chart
plt.figure(figsize=(8, 5))
plt.bar(value_counts.index.astype(str), value_counts.values, color='green', edgecolor='black')
plt.title('Frequency Distribution')
plt.xlabel('Sum')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
value_counts = df['exact_label'].value_counts().sort_index()

# Plot bar chart
plt.figure(figsize=(8, 5))
plt.bar(value_counts.index.astype(str), value_counts.values, color='skyblue', edgecolor='black')
plt.title('Frequency Distribution')
plt.xlabel('Label')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
